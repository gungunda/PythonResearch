```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from dateutil import parser
%matplotlib inline

df= pd.read_csv('./RI_1day.txt',header=None, 
                     names=['date','time','o','h','l','c','v'],
                     parse_dates=[['date', 'time']], 
                     index_col='date_time')
df['return']=df['c'].pct_change()*100
df=df.drop(df.index[0])
```

```python
df
```

```python
df['num'] = np.arange(0,len(df.index))
df['std'] = ''
df
```

```python
def calc(dt, frame=5, returns=[]):
    cur_ix = int(dt['num'])
    if cur_ix >= frame:
        return np.std(returns[cur_ix - frame:cur_ix])
df['std']=df.apply(calc, axis=1,frame=5,returns=df['return'].values)
```
