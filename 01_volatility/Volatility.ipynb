```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from dateutil import parser
%matplotlib inline

df= pd.read_csv('./RI_1day.txt',header=None, 
                     names=['date','time','o','h','l','c','v'],
                     parse_dates=[['date', 'time']], 
                     index_col='date_time')
df['return']=df['c'].pct_change()*100
df=df.drop(df.index[0])
```

```python
#df
```

```python
def calc(dt, frame=5, returns=[]):
    cur_ix = int(dt['num'])
    if cur_ix >= frame:
        return np.std(returns[cur_ix - frame:cur_ix])
df['num'] = np.arange(0,len(df.index))
df['std'] = ''
df['std']=df.apply(calc, axis=1,frame=22,returns=df['return'].values)
```

```python
#df.describe()
```

```python
df = df['2014-07':]
```

```python
def smooth(x,window_len=11,window='hanning'):
        if x.ndim != 1:
                raise ValueError, "smooth only accepts 1 dimension arrays."
        if x.size < window_len:
                raise ValueError, "Input vector needs to be bigger than window size."
        if window_len<3:
                return x
        if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
                raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
        s=np.r_[2*x[0]-x[window_len-1::-1],x,2*x[-1]-x[-1:-window_len:-1]]
        if window == 'flat': #moving average
                w=np.ones(window_len,'d')
        else:  
                w=eval('np.'+window+'(window_len)')
        y=np.convolve(w/w.sum(),s,mode='same')
        return y[window_len:-window_len+1]
    
smoothed = pd.Series(smooth(df['std'],60), index=df.index)

plt.figure(1,figsize=[18,18])
plt.subplot(211)
df['c'].plot()

plt.subplot(212)
df['std'].plot()
smoothed.plot(color='r')



```
